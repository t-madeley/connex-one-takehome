## Importing data


- We found some issues with the quotation marks and delimiters in the csv file, pandas was not able to effectively read the data due to some delimiters being inside quotes and some outside eg: `"56;""housemaid";`
- We defined a quick funciton to iterate over the rows, split on the delimiter irrespective of quotes and then remove the quotes.
- This resulted in all of the data being read as objects (strings), we used the provided data dict to correct the known dtypes then inspected the remaining features and added an appropriate dtype
- We have ~41K rows with 20 features which matches the expectation in the 'bank-additional-full.csv' from the UCI ML Repo

````python
def remove_quotes_and_split_row(row: str, delim: str) -> List[str]:
    """
    Clean, make lower case and split a row of data separated by the delimiter.

    Parameters
    ----------
    row : str
        A string representing a row of data.
    delim: str
        The delimiter to split on

    Returns
    -------
    List[str]
        A list containing the cleaned and split values from the row.
    """
    cleaned_row = row.split(delim)
    cleaned_row = [row_value.lower().replace('"', "") for row_value in cleaned_row]
    return cleaned_row
````

````python
# Read the CSV file
with open(DATA_PATH, "r") as file:
    # Read the header row separately and clean it
    header_row = next(file).strip()
    cleaned_header = remove_quotes_and_split_row(header_row, delim=";")
    # Read each line, clean it, and store it in a list
    cleaned_data = [remove_quotes_and_split_row(row.strip(), delim=";") for row in file]


data = pd.DataFrame(cleaned_data, columns=cleaned_header)

display(data)
data.describe()
````

### Checking for missing values

- We write a quick function to check for missing values (null values) as well as "unknown" values as identified.
- Similarly check for '-1' values as indicated in the data dictionary for the 'pdays' field.
- We found no null values
- Many fields do contain "unknown" values as previously noted.
    - We propose to leave these in as "unknown" values may be non random:
        - An "unknown" status for the "default" may indicate a customer with no financial history which may be higher risk
        - An "unknown" job status may be indicative of either very high or very low income individuals who prefer anonymity
        - Obviously these are quite large assumptions and in reality would discuss this would domain experts.
        - We will be careful to choose appropriate modelling approaches that can handle these and self feature select to some degree.
- We found no -1 values in the **'pdays'** when the data dictionary suggests we *should*.
    - We investigated further and found many 999 values instead.
    - We will convert these to null values to properly represent that they are 'missing'
    - There is an arugment for dropping the feature but the small number of actual values may have some signal for our model. 
    - Again this is a big assumption that would be discussed with a domain expert\
    - In the plots we can see the small number of remaining values seem to follow a logical distribution.

````python
def check_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Check for missing values and count values that are "unknown" in a DataFrame.

    Parameters:
    - df (DataFrame): The DataFrame to check.

    Returns:
    - missing_values (DataFrame): A DataFrame containing the count and percentage of NA values, 'unknown' values as well as the count of -1 values, for each column.
                                  If a field does not appear, all values were 0 (no missing, 'unknown' or -1 values)
    """
    # Calculate counts and percentages
    total_missing = df.isnull().sum()
    percent_missing = (total_missing / len(df)) * 100
    unknown_count = (df == "unknown").sum()
    percent_unknown = (unknown_count / len(df)) * 100
    minus_one_count = (df == "-1").sum()
    percent_minus_one = (minus_one_count / len(df)) * 100

    # Create a DF
    missing_values = pd.concat(
        [
            total_missing,
            percent_missing,
            unknown_count,
            percent_unknown,
            minus_one_count,
            percent_minus_one,
        ],
        axis=1,
        keys=[
            "Total_NA_Values",
            "Percent_NA_Values",
            "Total_Unknown_Values",
            "Percent_Unknown_Values",
            "Total_Minus_One_Values",
            "Percent_Minus_One_Values",
        ],
    )
    # Filtering rows with all 0 values for brevity
    missing_values = missing_values[
        (missing_values["Total_NA_Values"] > 0)
        | (missing_values["Total_Unknown_Values"] > 0)
        | (missing_values["Total_Minus_One_Values"] > 0)
    ].sort_values(
        by=["Total_NA_Values", "Total_Unknown_Values", "Total_Minus_One_Values"],
        ascending=False,
    )
    return missing_values
````

##  Bivariate analysis


### Categorical Feautres

- This is slightly trickier with a categorical target (we might use correlations etc for a regression task).
- To get a more impirical understanding of the relationship between a categorical variable we performed a chi2 independence test.
- This gives a statistical measure of whether there is a significant association between two categorical variables. 
- We set a significance level of alpha = 0.05
- Our null hypothesis is "There is no significant association between the categorical variables"
- Our alternative hypothesis is "There is a significant association betweent he categorical variables"

### Learnings
- Despite the researchers conducting comprehensive feature selection, it appears that **loan** and **housing** do not have a significant association to the target variable. This suggest they may not be good features. 
- All of the remaining features had extremely low p values, indicating a strong association - Meaning they may be powerful features.
- The temporal features seem to have a strong association. A crosstabulation and a stacked bar was plotted to try and visualise this change.
    - While **day_of_week** was very hard to visualise, it's somewhat clear in the **month**  variable.
    - We can see that in March, September and October we have proportionally more 'yes' values than in other months.
    - This all suggests there are some short and long term seasonal trends.
    - We will persist with not using a time windowed training/evaluation due to the lack of complete dates and to save time.

````python
# Find the categorical columns, removing the target itself
categorical_vars = (
    data.select_dtypes(include=["object", "category"]).drop(columns=["y"]).columns
)

# Perform chi-square test for each categorical variable
for var in categorical_vars:
    contingency_table = pd.crosstab(data[var], data["y"])
    chi2, p_value, _, _ = chi2_contingency(contingency_table)
    print(f"Chi-square test result for {var}: Chi2 = {chi2}, p-value = {p_value}")
    if p_value <= 0.05:
        print(f"'{var}' was found to have a significant association with the target! ")
    else:
        print(
            f"'{var}' was NOT found to have a significant association with the target."
        )
````

````python
# Convert 'month' and 'day_of_week' to categorical variables with appropriate order
month_order = [
    "jan",
    "feb",
    "mar",
    "apr",
    "may",
    "jun",
    "jul",
    "aug",
    "sep",
    "oct",
    "nov",
    "dec",
]
data["month"] = pd.Categorical(data["month"], categories=month_order, ordered=True)

day_order = ["mon", "tue", "wed", "thu", "fri"]
data["day_of_week"] = pd.Categorical(
    data["day_of_week"], categories=day_order, ordered=True
)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot for 'month'
cross_tab_month = pd.crosstab(data["month"], data["y"], normalize="all")
cross_tab_month.plot(kind="bar", stacked=True, ax=axes[0])
axes[0].set_title("Month vs. Target Variable (y)")
axes[0].set_xlabel("Month")
axes[0].set_ylabel("Count")

# Plot for 'day_of_week'
cross_tab_day_of_week = pd.crosstab(data["day_of_week"], data["y"], normalize="all")
cross_tab_day_of_week.plot(kind="bar", stacked=True, ax=axes[1])
axes[1].set_title("Day of Week vs. Target Variable (y)")
axes[1].set_xlabel("Day of Week")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
````

## Intriguing discovery

- Many unsuccessful calls have quite long durations, much longer than the mean successful call.
- This represents quite a large waste of call operator's time, it's preferable to get a quick no than a very long a drawn out one.
- To explore this further we plotted the distribution, splitting by 'no' and 'yes' outcomes.
- We calculated the the percentage of 'no' calls with a duration greater than the mean 'yes' call and found that only 6% of unsuccesful calls last longer than a the mean successful one.
- While this seems small, we thought it may represent a signficant proportion of a call operators time.
- We calculated the percentage of the overall 'no' call duration that exceeds the 'yes' call mean duration and found that this is actually 22% of the 'no' call time, which is huge.
- This is interesting because it suggests that there are large time savings to be had my optimising the 'no' call duration.
- You could do this with simple statistical methods like modelling the propbability of call success as duration increases or more complex models like audio or transcript analysis, looking for key words or phrases that would suggest and unsuccessful call earlier.
- Similarly training prompts for the call operator could be updated as duration increases, either pushing the call operator to complete the call successfully or terminate it and move on.
- I just thought it was interesting! 

````python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the mean duration where 'y' equals 'yes'
mean_duration_yes = data[data["y"] == "yes"]["duration"].mean()

# Plot the distribution of 'duration' for each group of 'y'
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x="duration", hue="y")
plt.axvline(
    x=mean_duration_yes, color="red", linestyle="--", label="Mean Duration (y=yes)"
)
plt.title("Distribution of Duration by Target Variable (y)")
plt.xlabel("Duration (Truncated for legibility)")
plt.ylabel("Frequency")
plt.legend()
plt.xlim(0, 2000)  # Limit the x-axis to 3000
plt.show()
````

````python
# Calculate the proportion of 'no' durations longer than the mean 'yes' call duration
proportion_no_longer_than_mean_yes = (
    data[(data["y"] == "no") & (data["duration"] > mean_duration_yes)].shape[0]
    / data[data["y"] == "no"].shape[0]
) * 100

print(
    f"Percentage of 'no' durations longer than the mean 'yes' call duration: {round(proportion_no_longer_than_mean_yes, 2)}%"
)
````

````python
# Calculate the total duration of 'no' calls longer than the mean 'yes' call duration
total_duration_no_longer_than_mean_yes = data[
    (data["y"] == "no") & (data["duration"] > mean_duration_yes)
]["duration"].sum()

total_calls_no_longer_than_mean_yes = data[
    (data["y"] == "no") & (data["duration"] > mean_duration_yes)
].shape[0]

# Calculate the total duration of all 'no' calls
total_duration_no = data[data["y"] == "no"]["duration"].sum()

# Calculate the percentage of the total duration represented by 'no' calls longer than the mean 'yes' call duration
percentage_total_duration = (
    total_duration_no_longer_than_mean_yes / total_duration_no
) * 100

print(
    f"Percentage of the total 'no' call duration represented by durations longer than the mean 'yes' call duration: {round(percentage_total_duration, 2)}%"
)
````

- As we are using models that are invariant to scale, we will not perform any data transformations


### Encoding Variables 

- We could just create an ordinal encoding for all variables, however this suggests to the model that there is an ordinal relationship when there may not be.
- We will use ordinal encoding when there is a logical, numerical progression to the values eg: "month" or "education".
    - **note** Ordinal encoding is not ideal for date features, we could improve of this using a sine and cosine transofrmation instead to better represent the cyclical nature 
- We will use one hot encoding for variables where there is no logical ordering.

````python
ordinal_variables = ["month", "day_of_week", "education", "poutcome", "contact", "y"]

one_hot_variables = ["job", "marital", "default"]
````

````python
from sklearn.preprocessing import OrdinalEncoder

# Define category orders, being explicit to preserve order.
category_orders = {
    "month": [
        "jan",
        "feb",
        "mar",
        "apr",
        "may",
        "jun",
        "jul",
        "aug",
        "sep",
        "oct",
        "nov",
        "dec",
    ],
    "day_of_week": ["mon", "tue", "wed", "thu", "fri"],
    "education": [
        "illiterate",
        "unknown",
        "basic.4y",
        "basic.6y",
        "basic.9y",
        "high.school",
        "professional.course",
        "university.degree",
    ],
    "poutcome": ["failure", "nonexistent", "success"],
    "contact": ["cellular", "telephone"],
    "y": ["no", "yes"],
}

# Create and apply OrdinalEncoders for each variable
ordinal_encoders = {
    col: OrdinalEncoder(categories=[order]) for col, order in category_orders.items()
}
data_encoded = data.copy()  # Make a copy of the original data

# Encode each variable using its corresponding encoder
for col, encoder in ordinal_encoders.items():
    data_encoded[col] = encoder.fit_transform(data_encoded[[col]])
````

````python
# One-hot encoding for one-hot variables
data_one_hot = pd.get_dummies(
    data[one_hot_variables], columns=one_hot_variables
).astype(int)
# Quick check that all three variables were one hot encoded correctly
assert all(data_one_hot.sum(axis=1) == 3)
````

### Joining Numerical, Ordinal and One Hot encoded features
- We join on the index to create one feature matrix
- We use pandas join 'validate' argument to ensure the join is 1:1

````python
encoded_data = data_one_hot.join(
    data_encoded.drop(one_hot_variables, axis=1), validate="1:1"
).drop("pdays", axis=1)
````

````python
display(encoded_data)
````

### Train Test Split

````python
X = encoded_data.iloc[:, :-1]
y = encoded_data.iloc[:, -1]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=123
)
````

````python
outlier_features = ["age", "duration", "campaign"]

# Calculate the 99th percentile for each numerical feature, careful only to use the training data
percentile_99 = X_train[outlier_features].quantile(0.99)
print(percentile_99)
# Remove outliers above the 99th percentile for each feature in both training and test sets
# Create boolean masks for filtering outliers in both training and test sets
train_outlier_mask = ~X_train[outlier_features].gt(percentile_99).any(axis=1)
test_outlier_mask = ~X_test[outlier_features].gt(percentile_99).any(axis=1)

# Filter both the features and the target variable based on the outlier masks
X_train = X_train[train_outlier_mask]
y_train = y_train[train_outlier_mask]
X_test = X_test[test_outlier_mask]
y_test = y_test[test_outlier_mask]
````

# Modelling

- We will create two tree based models as they are invariant to feature scales and are generally deal with categorical features well:
1. RandomForestClassifier
    - This represents a bootstrap aggregation approach where many decision trees are trained in parallel using bootstrapped samples of our dataset and outputs are averaged.
    - This can help reduce variance and overfitting.
2. Gradient Boosted Trees Classifier
    - This represents a boosting approach where many trees are trained sequentially, at each step the errors are penalized before the next model is trained.
    - GB trees use gradient decent to tweak the penalties at each step.

For both models we will perform a randomised cross validated parameter search over a small grid. This will randomly sample parameters from our grid to find the optimum hyperparameter set, optimising for maximum ROC AUC.

## Random Forest Model

````python
# Define the parameter grid for RandomizedSearchCV
param_dist = {
    "n_estimators": randint(50, 200),
    "max_depth": randint(5, 20),
    "min_samples_split": randint(2, 20),
    "min_samples_leaf": randint(1, 10),
    "criterion": ["gini", "entropy"],  # Add criterion parameter
}


rf_classifier = RandomForestClassifier(random_state=123)

# Initialize RandomizedSearchCV with 5-fold cross-validation
randomized_search = RandomizedSearchCV(
    estimator=rf_classifier,
    param_distributions=param_dist,
    n_iter=20,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
    random_state=123,
)

# Perform the random search on the training data
randomized_search.fit(X_train, y_train)

# Get the best parameters and best accuracy
best_params = randomized_search.best_params_
best_roc_auc = (
    randomized_search.best_score_
)  # Just to make sure model is learning, will evaluate properly later

print("Best Parameters:", best_params)
print("Best ROC AUC:", best_roc_auc)

# Initialize the Random Forest Classifier with the best parameters
best_model = RandomForestClassifier(**best_params, random_state=123)

# Train the best model on the full training data
best_model.fit(X_train, y_train)
````

## Gradient Boosted Trees Model

````python
# Define parameter grid
param_grid = {
    "n_estimators": randint(50, 200),
    "learning_rate": [
        0.01,
        0.05,
        0.1,
        0.2,
        0.5,
    ],  # Will not use random values for learning rate as this needs to be more precise
    "max_depth": randint(3, 10),
    "min_samples_split": randint(2, 20),
    "min_samples_leaf": randint(1, 10),
}

# Initialize Gradient Boosting Classifier
gb_classifier = GradientBoostingClassifier(random_state=123)

# Initialize RandomizedSearchCV
random_cv_gb = RandomizedSearchCV(
    estimator=gb_classifier,
    param_distributions=param_grid,
    n_iter=20,
    scoring="roc_auc",
    cv=5,
    n_jobs=-1,
    random_state=123,
)

# Fit RandomizedSearchCV
random_cv_gb.fit(X_train, y_train)

# Get best estimator
best_gb_estimator = random_cv_gb.best_estimator_
# Get the best parameters and best accuracy
best_params_gb = random_cv_gb.best_params_
best_roc_auc_gb = random_cv_gb.best_score_
print("Best Parameters:", best_params_gb)

print("Best ROC AUC ", best_roc_auc)

best_model_gb = GradientBoostingClassifier(**best_params_gb, random_state=123)

best_model_gb.fit(X_train, y_train)
````

# Model Evaluation

We built two functions to create some basic plots of the model metrics
1. Calculates Precision, Recall and F1 Score
2. Calculates the ROC Curve. 

````python
from sklearn.base import BaseEstimator


def calculate_metrics(
    model: BaseEstimator, X_test: pd.DataFrame, y_test: pd.Series
) -> dict:
    """Calculate basic metrics for binary classification model.

    Parameters
    ----------
    model: The fitted sklearn base estimator using optimum parameters
    X_test: The test feature set
    y_test: The test labels

    Returns
    -------
    dictionary of metrics
    """
    # Predict the target variable for the test set
    y_pred = model.predict(X_test)

    # Calculate precision, recall, f1 score
    precision, recall, f1_score, _ = precision_recall_fscore_support(
        y_test, y_pred, average="binary"
    )

    metrics = {"Precision": precision, "Recall": recall, "F1 Score": f1_score}

    return metrics


def calculate_roc_auc(model, X_test, y_test) -> None:
    """Calculate the ROC AUC the corresponding ROC Curve for binary classification model.

    Parameters
    ----------
    model: The fitted sklearn base estimator using optimum parameters
    X_test: The test feature set
    y_test: The test labels

    Returns
    -------

    """
    # Predict the target variable probabilities for the test set
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    # Calculate ROC AUC
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Calculate false positive rate, true positive rate, thresholds
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

    metrics = {
        "ROC_AUC": roc_auc,
        "False Positive Rate": fpr,
        "True Positive Rate": tpr,
        "Thresholds": thresholds,
    }

    return metrics
````

**Comparison**:
- Precision: Random Forest outperforms Gradient Boosted Trees slightly
- Recall: Gradient Boosted Trees have a higher recall by about 0.12 compared to Random Forest.
- F1 Score: Despite having lower precision, Gradient Boosted Trees achieve a slightly higher F1 score than Random Forest.
- ROC AUC: The Gradient Boost Trees had a slightly higher ROC AUC score, meaning it has slightly better performance across more classification thresholds.
- The feature importances for the models were almost identical for the features with high importance.
    - The key difference is the Gradient Boosted Trees model valued the **nr.employed** feature much more highly.
    - The Random forest model also put more importance on the **job** feature, where the Gradient boosted trees model almost completely ignored them.


**Precision**: Random Forest may achieve higher precision because it's less prone to overfitting, resulting in fewer false positives. If we had more time, we would dig in to the training accuracies and look for overfitting. 
**Recall**: Gradient Boosted Trees might have higher recall due to its iterative nature, allowing it to focus on difficult-to-classify instances, leading to fewer false negatives.
**F1 Score**: Gradient Boosted Trees strike a better balance between precision and recall, possibly due to its iterative approach, which fine-tunes predictions with each new model iteration.
**ROC AUC**: Similarly Gradient Boosted Trees has a slightly higher ROC AUC due to the better balance between precision and recall.

**Conclusion**

- The Gradient Boost Trees model *slightly* outperformed a random forest model in terms of ROC AUC but the performance difference was extremely close.
- Ultimately the choice between models is not clear cut however.
- We should speak to domain experts and find out their priorities:
    - Do they require a highly precise model?
        - This would minimise wasted call time by focusing on only the customers the model is most confident will make a purchase.
    - Do they require a high recall model capable of finding more potential customers?
        - This could potentially maximise overall revenue by finding more customers that are willing to make purchases at the expense of more wasted call time (because it is less precise).
     
- Interestingly, both of these models were slightly better than the NN approach used in the paper when NOT using the rolling window evaluation method: Paper NN scored 0.929 ROC AUC. The difference is very small and could simply be due to sampling differences though.
